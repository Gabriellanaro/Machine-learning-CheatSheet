```python
import numpy as np


def calculate_rss_impurity(values):
    mean_value = np.mean(values)
    rss = np.sum((values - mean_value) ** 2)
    return rss / len(values)


def purity_gain_rss1(nodes):
    """
    Calculates the purity gain using RSS (Residual Sum of Squares) for a given split.

    Parameters:
    - nodes (list of numpy.ndarray): List containing values of nodes after the split.

    Returns:
    - float: Purity gain using RSS.
    """

    rss_values = []
    total_instances = sum(len(node) for node in nodes)

    for node_values in nodes:
        rss_node = calculate_rss_impurity(node_values)
        rss_values.append(rss_node)
        print(f"I(v_k): {rss_node}")

    purity_gain = rss_values[0] - sum(
        (len(node) / total_instances) * rss for node, rss in zip(nodes, rss_values[1:])
    )

    return purity_gain
```

```python
parent_values = np.array([-1.76, 0, 0.06, 0.08, 0.65, 1.3])  # Values of the parent node
left_child_values = np.array(
    [-1.76, 0, 0.06, 0.08]
)  # Values of the left child node after the split
right_child_values = np.array(
    [0.65, 1.3]
)  # Values of the right child node after the split

purity_gain = purity_gain_rss1([parent_values, left_child_values, right_child_values])
print(f"Purity Gain using RSS: {purity_gain}")
```
